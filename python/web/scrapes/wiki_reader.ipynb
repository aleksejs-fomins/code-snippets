```python
import numpy as np
import requests
from PIL import Image
from urllib.request import urlopen
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
from IPython.core.display import display, HTML
```

```python
url = "https://en.wikipedia.org/wiki/Segnosaurus"

r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
text = soup.get_text()
```

```python
imgs = soup.find_all('img')
for img in imgs:
    caption = img.get('alt')
    imgurl = img.get('src')
    if imgurl[:2] == '//':
        plt.figure()
        plt.imshow(Image.open(urlopen("https:"+imgurl)))
        plt.title(caption)
        plt.axis('off')
        plt.show()
```
