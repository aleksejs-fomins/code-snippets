```python
%matplotlib notebook
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

class MyPlot:
    def __init__(self, w=1.0):
        self.fig, self.ax = plt.subplots(figsize=(4,4))
        
        self.x = np.linspace(0, 2 * np.pi)
        self.line, = self.ax.plot(self.x, self.f(w))
        
    def f(self, w):
        return np.sin(w * self.x)

    def update(self, w = 1.0):
        self.line.set_ydata(self.f(w))
        self.fig.canvas.draw_idle()
        
myPlot1 = MyPlot(1.0)
interact(myPlot1.update);
```
